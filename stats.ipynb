```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("DemocracyCSV.csv")

plt.figure(4)
plt.clf()
# Afghanistan
inds, =  np.where(data['Entity'] == 'Afghanistan')
bins = np.linspace(0,1,25)
plt.hist(data["electdem_vdem_owid"][inds].values, bins=bins,label='Afghanistan',alpha=0.5)

vdem_array = data["electdem_vdem_owid"][inds].values
vdem_array = vdem_array[~np.isnan(vdem_array)]
print('V_DEM for Adghanistan: ')
print('Mean: ' + str(np.mean(vdem_array)))
print('Median: ' + str(np.median(vdem_array)))
print('Std: ' + str(np.std(vdem_array)) + '\n')


# Sweden
c = 'Sweden'
inds, =  np.where(data['Entity'] == c)
plt.hist(data["electdem_vdem_owid"][inds].values, bins=bins,label=c,alpha=0.5)

vdem_array = data["electdem_vdem_owid"][inds].values
vdem_array = vdem_array[~np.isnan(vdem_array)]
print('V_DEM for Sweeden: ')
print('Mean: ' + str(np.mean(vdem_array)))
print('Median: ' + str(np.median(vdem_array)))
print('Std: ' + str(np.std(vdem_array)))


#print(np.mean(data["electdem_vdem_owid"][inds].values))
plt.legend()
```
